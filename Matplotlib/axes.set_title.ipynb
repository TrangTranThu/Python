```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(-10, 10, 0.1)
b = 3*a**3 + 2*a**2 + 3*a + 1
```

```python
fig, ax = plt.subplots()
ax.set_title('ĐỒ THỊ 1')
ax.plot(a, b);
```

```python
fig, ax = plt.subplots()
ax.set_title('ĐỒ THỊ 2')
ax.set_xlabel('Truc x', loc='right')
ax.set_ylabel('Truc y', loc='top')
ax.plot(a, b);
```
