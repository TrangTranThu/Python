# Introduction to pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4]);
plt.ylabel('some numbers');
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 7], [4, 2]);
```

# Formatting the style of your plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 10], 'b-'); # mạc định là 'b-': là đường màu xanh lam
```

```python
plt.plot([1, 2, 3], [1, 4, 5], 'ro');
plt.axis([0, 6, 0, 15]); # [xmin, xmax, ymin, ymax]
```

```python
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', t, t**t, 'bs', t, t**3, 'r^');
```
