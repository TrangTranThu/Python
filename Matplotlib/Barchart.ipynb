## Axes.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
a = np.array([11, 20, 33, 40, 25])
b = ['dog', 'cat', 'bird', 'duck', 'fish']
ax.set_title('ANIMAL', color='red', size=15)
ax.set_xlabel('animal', loc='right', color='red')
ax.set_ylabel('values', loc='top', color='red')
ax.axis([-1, 5, 0, 50])
ax.bar(b, a);
```
