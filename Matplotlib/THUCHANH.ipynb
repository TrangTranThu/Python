* Vẽ hai biểu đồ ở trên trong cùng một cửa sổ 


```python
import numpy as np
import matplotlib.pyplot as plt
```

x = np.array([1, 3, 5, 7, 10, 12, 24, 25, 30, 40])
y = np.array([3, 2, 6, 4, 8, 5, 3, 8, 7, 1])

a = np.arange(-10, 10, 0.1)
b = 3*a**3 - 3*a**2 + 3*a - 3

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y);
ax2.plot(a, b);
