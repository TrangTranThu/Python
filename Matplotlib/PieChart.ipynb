```python
import numpy as np
import matplotlib.pyplot as plt
```

# x

```python
x = np.array([3, 12, 5, 8])
```

```python
# x : 1D array-like
#     The wedge sizes.
plt.pie(x);
```

# explode

```python
#explode: array-like, optional, default: None
# If not None, 
# is a len(x) array which specifies the fraction of the radius with which to offset each wedge.
# plt.pie(x, explode = [0.1, 0, 0, 0]);
# plt.pie(x, explode = [0, 0.2, 0, 0]);
plt.pie(x, explode = [0.2, 0, 0.1, 0]);
```

# labels

```python

#labels : list, optional, default: None
# A sequence of strings providing the labels for each wedge
labels = ['Trang', 'Hong', 'Hanh', 'Nga'];
# plt.pie(x, labels = y);
plt.pie(x, labels = labels);
```

# color

```python
# colors : array-like, optional, default: None
# A sequence of matplotlib color args through which the pie chart will cycle. If None, will use the colors in the currently active cycle.

plt.pie(x, colors = ['black', '#4CAF50', 'red', 'blue']);
```

# autopct

```python
# autopct (%): None (default), string, or function, optional
# If not None, is a string or function used to label the wedges with their numeric value. The label will be placed inside the wedge. If it is a format string, the label will be fmt%pct. If it is a function, it will be called.
 
plt.pie(x, labels=['a', 'b', 'c', 'd'] , autopct='%.2f');
```

```python
values = [10, 200, 33, 44]

plt.pie(values, labels=labels, autopct=lambda p : '{:.1f}% ({:,.0f})'.format(p,p * sum(values)/100));
```

```python
# make the pie circular by setting the aspect ratio to 1
plt.figure(figsize=plt.figaspect(1))
values = [3, 12, 5, 8] 
labels = ['a', 'b', 'c', 'd'] 

def make_autopct(values):
    def solve(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f} %  ({v:d})'.format(p=pct,v=val)
    return solve

plt.pie(values, labels=labels, autopct=make_autopct(values));
```

## pctdistance : float, optional, default: 0.6
* The ratio between the center of each pie slice and the start of the text generated by autopct. Ignored if autopct is None.

```python
labels = ['python', 'C++', 'java', 'golang']
sizes = [215, 130, 245, 2000]
ex = (0.1, 0, 0, 0)  # explode 1st slice
plt.pie(sizes, labels=labels, autopct='%1.2f%%', pctdistance=0.65, startangle=140);
```

```python
import matplotlib.pyplot as plt

fig =plt.figure(figsize = (10,10))
ax11 = fig.add_subplot(111)
# Data to plot
sizes = [215, 130, 245, 2000]
labels = 'Python', 'C++', 'Ruby', 'Java'
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
w,l,p = ax11.pie(sizes,  labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=140, pctdistance=0.65, radius = 1, explode=explode)
[t.set_rotation(0) for t in p]
[t.set_fontsize(25) for t in p]
[t.set_fontsize(25) for t in l]
plt.axis('equal')
plt.show()
```

# shadow : bool, optional, default: False
* Draw a shadow beneath the pie.

```python
plt.pie(sizes, labels=labels, autopct='%1.2f%%', pctdistance=0.65, shadow=True);
```

# labeldistance : float or None, optional, default: 1.1
* The radial distance at which the pie labels are drawn. If set to None, label are not drawn, but are stored for use in legend()

```python
plt.pie(sizes, labels=labels, autopct='%1.2f%%', pctdistance=0.65, shadow=True, labeldistance = 1.2);
```

# startangle : float, optional, default: None
* If not None, rotates the start of the pie chart by angle degrees counterclockwise from the x-axis.

```python
plt.pie(sizes, labels=labels, autopct='%1.2f%%', pctdistance=0.65, shadow=True, labeldistance = 1.2 , startangle = 90);
```

# radius : float, optional, default: None
* The radius of the pie, if radius is None it will be set to 1.

```python
plt.pie(sizes, labels=labels, autopct='%1.2f%%', pctdistance=0.65, shadow=True, labeldistance = 1.2 , startangle = 90, radius = 1);
```

# frame : bool, optional, default: False
* Plot axes frame with the chart if true.

```python
plt.pie(sizes, labels=labels, autopct='%1.2f%%', pctdistance=0.65, labeldistance=1, frame=True);
```

# center : list of float, optional, default: (0, 0)
* Center position of the chart. Takes value (0, 0) or is a sequence of 2 scalars.

```python
plt.pie(sizes, labels=labels, autopct='%1.2f%%', pctdistance=0.65, labeldistance=1, startangle = 90, frame=True, center=(0.5,0.5),);
```
