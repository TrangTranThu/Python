## loc vs iloc

### địa chỉ, index location(vị trí theo địa chỉ)

```python
df.head(5)
```

```python
df.loc[df.quantity == 15] 
```

```python
df.loc[df.item_name == 'Nantucket Nectar']
```

```python
df.loc[(df.quantity == 9) | (df.item_name == 'Nantucket Nectar')] 
```

```python
# select all rows with multiple conditions
df.loc[(df.quantity >=2) & (df.item_name == 'Nantucket Nectar'), ['order_id', 'quantity']]
# ['order_id']: để hiển thị những cột muốn hiển thị
```

### Data Manipulation

```python
df.item_price.dtype
```

### Apply()

```python
df.item_price = df.item_price.apply(lambda x : float(x.replace('$', '')))
```

```python
df.head()
```

```python
df['total_price'] = df['quantity'] * df['item_price']
```

```python
df.head()
```

```python
revenue = df['total_price'].sum()
```

```python
revenue
```

## groupby

```python
c = df.groupby('item_name')['quantity'].sum()
```

```python
c.sort_values() # mặc định là tăng dần
```

```python
c.sort_values(ascending = False).head(5)
```
